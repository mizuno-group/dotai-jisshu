```python
# @title 物質収支式の入力（こちらは流石に長いので、共有スプシに分担入力後6行分のセルをこの下に**直接**コピペ、さらに funcs= と挿入してから実行してください。）
"""
import io
import pandas as pd
pd.read_csv(io.StringIO('''
...
このように表示されていればコピペは成功です。
pd.read_csvの前に funcs= と挿入してから、実行してください。
"""
# ↓↓↓ コピペ
```

```python
# @title 物質収支式に必要なパラメータの入力（こちらも流石に長いので、共有スプシに分担入力後18行8列 or 18行6列のセルを**直接**コピペ、さらに params= と挿入してから実行してください。）
"""
import io
import pandas as pd
pd.read_csv(io.StringIO('''
...
このように表示されていればコピペは成功です。
pd.read_csvの前に params= と挿入してから、実行してください。
"""
# ↓↓↓ コピペ
```

```python
# @title 読み込みデータの確認・整形（そのまま実行してください。）
assert "funcs" in locals(), "funcsが定義されていません。物質収支式のコピペを行うセルに戻って指示に従ってください。"
assert "params" in locals(), "paramsが定義されていません。物質収支式に必要なパラメータのコピペを行うセルに戻って指示に従ってください。"
assert funcs.shape == (6, 1), "funcsの形状が6行1列ではありません。物質収支式のコピペを行うセルに戻って指示に従ってください。"

compiled_funcs = [
    compile(str(funcs.iloc[i, 0]), f"func_{i}", "eval")
    for i in range(6)
]
def PBPK(t, c, p):
    local_vars = {
        "t": t,
        "c": c,
        "p": p,
    }
    return np.array([
        eval(expr, {"__builtins__": {}}, local_vars)
        for expr in compiled_funcs
    ], dtype=float)

if group == "奇数班":
    assert params.shape == (18, 6), f"奇数班：paramsの形状が18行6列ではありません。現在は{params.shape}という形状です。物質収支式に必要なパラメータのコピペを行うセルに戻って指示に従ってください。"
    params.columns = pd.MultiIndex.from_tuples([
        ("HX1", "ctrl"), ("HX1", "fB*3"), ("HX1", "Qh*2"),
        ("LX1", "ctrl"), ("LX1", "fB*3"), ("LX1", "Qh*2"),
    ])
elif group == "偶数班":
    assert params.shape == (18, 8), f"偶数班：paramsの形状が18行8列ではありません。現在は{params.shape}という形状です。物質収支式に必要なパラメータのコピペを行うセルに戻って指示に従ってください。"
    params.columns = pd.MultiIndex.from_tuples([
        ("HX1", "ctrl"), ("HX1", "PSinf*0.5"), ("HX1", "CLbile*0.5"), ("HX1", "CLbile*0.1"),
        ("LX1", "ctrl"), ("LX1", "PSinf*0.5"), ("LX1", "CLbile*0.5"), ("LX1", "CLbile*0.1"),
    ])

display(params)
print("データの確認・整形が完了しました。次のセルに進んでください。")
```

```python
# @title HX1について1000分までシミュレーション（そのまま実行してください。ある程度時間がかかります。）
params_hx1 = params["HX1"]
if group == "奇数班":
    p1 = tuple(params_hx1["ctrl"])
    p2 = tuple(params_hx1["fB*3"])
    p3 = tuple(params_hx1["Qh*2"])
    runge_kutta(PBPK, [p1,p2,p3], ["ctrl","fB*3","Qh*2"])
elif group == "偶数班":
    p1 = tuple(params_hx1["ctrl"])
    p2 = tuple(params_hx1["PSinf*0.5"])
    p3 = tuple(params_hx1["CLbile*0.5"])
    p4 = tuple(params_hx1["CLbile*0.1"])
    runge_kutta(PBPK, [p1,p2,p3,p4], ["ctrl","PSinf*0.5","CLbile*0.5","CLbile*0.1"])
```
